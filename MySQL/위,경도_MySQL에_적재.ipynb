```python
values_list = []

query = insert(table)

for a in df['지점번호']:
  b = df.loc[df['지점번호'] == a, '지점명'].values[0]
  c = df.loc[df['지점번호'] == a, '위도'].values[0]
  d = df.loc[df['지점번호'] == a, '경도'].values[0]
  e = df.loc[df['지점번호'] == a, '관리관서'].values[0]

  values_dict = {'stnld' : a, 'stnNm' : b, 'mngdp' : e, 'lat' : c, 'lgt' : d}
  values_list.append(values_dict)

result_proxy = connection.execute(query, values_list)
result_proxy.close()
```
