```python
result_list=[]
for i in range(3646, 4632):
    try:
        df = pd.read_csv(FILE_COMMON_NAME+str(i)+'.csv', encoding='cp949')
        row = list(df.index)
        for k in range(2,len(row)):
            result_list.append(row[k])
        try:
            df = pd.read_csv(FILE_COMMON_NAME+str(i)+' (1).csv', encoding='cp949')
            row = list(df.index)
            for k in range(2,len(row)):
                result_list.append(row[k])
        except:
            continue
        
    except:
        continue
    
result_list
```

```python
df_result = pd.DataFrame(result_list, columns = ['지점번호', '지점명','일시', '실효습도'])
df_result
```

```python
df_result.to_csv('new_sample_humid.csv')
```

```python
# merge final weather history
FILE_COMMON_NAME = 'final_weather_2004_2011_'
```

```python
result_df = pd.read_csv(FILE_COMMON_NAME + '0' +'.csv', low_memory=False)

for i in range(1,9):
    temp_df = pd.read_csv(FILE_COMMON_NAME + str(i) +'.csv', low_memory=False)
    result_df = pd.concat([result_df,temp_df], ignore_index=True)

result_df
```
